
## Populate the sales data from parquet to sql table

In the lab **Azure Synapse Analytics and AI hands-on lab step-by-step**, **Exercise 2** has showed how to  create and populate the supporting tables in the SQL Pool.

Now, this notebook will demostrate how to perform the similar task using Spark. Here will translate **Exercise 2:Task 1: Create the sale table** to spark practic


## Load 2019 sales data and explore the data


```scala
var sales_2019_df =spark.read.parquet("abfss://wwi-02@asadatalakewy.dfs.core.windows.net/sale-small/Year=2019/sale-small/Year=2019/Quarter=Q1/Month=1")
sales_2019_df.createOrReplaceTempView("viewSales2019")
```

```scala
display(sales_2019_df.limit(10))
```

```scala
display(spark.sql("select * from viewSales2019"))
```

## Column Mapping 
![Image alt text](https://github.com/wangyihaier/MCW-Azure-Synapse-Analytics-and-AI/raw/master/Hands-on%20lab/media/dataflow_sink_mapping.png)


```scala
var sqlQuery = s"""Select 
    TransactionId,
    CustomerId,
    ProductId,
    int(Quantity) as Quantity,
    Price,
    TotalAmount,
    TransactionDate as TransactionDateId,
    ProfitAmount,
    int(Hour) as Hour,
    int(Minute) as Minute,
    int (StoreId) as StoreId
From viewSales2019"""

var small_sales_df = spark.sql(sqlQuery)
//small_sales_df.cache()
```

```scala
display(small_sales_df.limit(10))
```

```scala
small_sales_df.printSchema()
```

## Sink the data to the sql table 
**This feature is only available in Scala so far**


```scala
small_sales_df.write.sqlanalytics("SQLPool01.wwi_mcw.SaleSmall2019",Constants.INTERNAL)
```

## Verify the result

```scala
val df = spark.read.sqlanalytics("SQLPool01.wwi_mcw.SaleSmall2019") 
df.show(10)
```
